```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load the dataset

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

## check null values

```python
df.isnull().sum()
```

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Scatter plot of Per Capita income over years')
plt.show()
```

## prepare the data for machine learning

```python
X = df.drop('year', axis=1)
y = df.drop('per capita income (US$)', axis=1)
```

```python
# Split the data into training and test sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 4)
```

```python
# create a linear regression model and fit it to training data

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# slope 
lr.coef_ 
```

```python
# intercept
lr.intercept_
```

```python
# prediction on test data

y_pred = lr.predict(X_test)
y_pred
```

```python
# models performance

lr_score = lr.score(X_test, y_test)
lr_score
```

```python
# plot training data and regression line

plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_train, lr.predict(X_train), color='red', label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.legend()
plt.title('Linear Regression Model')
plt.show()
```
